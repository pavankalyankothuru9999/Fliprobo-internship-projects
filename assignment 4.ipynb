# Scrape the details of State-wise GDP of India from statisticstime.com

```python
#importing the libraries
import selenium
import pandas as pd
from selenium import webdriver
# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
#importing the driver
driver =webdriver.Chrome(r"C:\Users\pavan\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
#importing the url
url = 'http://statisticstimes.com/economy/india/indian-states-gdp.php'
driver.get(url)
```

```python
#scraping the rank tag
rank_tags=driver.find_elements_by_xpath("//td[@class='data1']")
#appling the forloop
rank=[]
for i in rank_tags:
    rank.append(i.text)
ranks=rank[0:33]
ranks
```

```python
#scraping the name tag
name_tags=driver.find_elements_by_xpath("//td[@class='name']")
#appling the forloop
name=[]
for i in name_tags:
    name.append(i.text)
names=name[0:33]
names
```

```python
#scraping the GSDP at current price (18-19) tag
name_tags=driver.find_elements_by_xpath("//td[@class='data sorting_1']")
#appling the forloop
GSDP_18=[]
for i in name_tags:
    GSDP_18.append(i.text)
GSDP_18_19=GSDP_18[0:33]
GSDP_18_19
```

```python
#scraping the name tag
a_tags=driver.find_elements_by_xpath("//td[@class='data']")
#appling the forloop
a=[]
for i in a_tags:
    a.append(i.text)
GSDP_19=a[0::5]
GSDP_19_20=GSDP_19[0:33]
GSDP_19_20
```

```python
shares=a[1::5]
share=shares[0:33]
share
```

```python
gdp=a[2::5]
GDP=gdp[0:33]
GDP
```

```python
GDP_INDIA=pd.DataFrame({})
GDP_INDIA['Rank']=ranks
GDP_INDIA['State']=names
GDP_INDIA['GSDP at current price (18-19)']=GSDP_18_19
GDP_INDIA['GSDP at current price (19-20)']=GSDP_19_20
GDP_INDIA['Share(18-19)']=share
GDP_INDIA['GDP($ billion)']=GDP
```
