```python
bos_new=bos[(z<3).all(axis=1)]     # Remove outlier
```

```python
bos=bos_new
bos.shape
```

```python
x=bos.iloc[:,0:-1]
x.head()
```

```python
y=bos.iloc[:,0:-1]
x.head()
```

```python
x.shape
```

```python
y.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.33,random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
lm=LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
lm.score(x_train,y_train)
```

```python
# predict the value
pred=lm.predict(x_test)
print("predicted result price:",pred)
print("actual price",y_test)
```

```python
print('error:')
print('Mean absolute error:',mean_absolute_error(y_test,pred))
print('Mean squared error:',mean_squared_error(y_test,pred))

print('Root Mean Squared error:',np.sqrt(mean_squared_error(y_test,pred)))
```

```python
# r2 score->coefficiant of determination
# is the proportion of variance in the dependend variable that is predictable from the independent variable
#(s) change comming in y whenever x is being charged
from sklearn.metrics import r2_score
print(r2_score(y_test,pred))
```
